# Task
Let's calculate daily PnL for each account by grouping trades by date and account, then see who the top performers are.

## Calculate Daily PnL per Account

### Subtask:
Break down profit and loss by account and date, then identify the accounts making the most money.


**Reasoning**:
I'll group the data by date and account, then add up all the closed PnL for each combination.



```python
daily_pnl_per_account = merged_df.groupby(['date', 'Account'])['Closed PnL'].sum().reset_index()
print("Daily PnL per account (first 5 rows):")
print(daily_pnl_per_account.head())
```

**Reasoning**:
Now I want to see who's really winning overall. I'll total up each account's PnL across all days and rank them from best to worst.



```python
total_pnl_per_account = daily_pnl_per_account.groupby('Account')['Closed PnL'].sum()
top_10_accounts = total_pnl_per_account.sort_values(ascending=False).head(10)

print("Top 10 accounts by total PnL:")
print(top_10_accounts)
```

## Calculate Win Rate and Average Trade Size

### Subtask:
Figure out what percentage of trades are profitable for each account, and see how big their average positions are.


**Reasoning**:
For each account, I need to count how many trades made money versus total trades, then turn that into a percentage.



```python
win_rate_per_account = merged_df.groupby('Account').apply(lambda x: (x['Closed PnL'] > 0).sum() / len(x) * 100)
print("Win Rate per Account (%):")
print(win_rate_per_account.sort_values(ascending=False).head())
```

**Reasoning**:
Let me clean this up with a better approach using .agg(), plus I still need to calculate average position sizes.



```python
win_rate_per_account = merged_df.groupby('Account')['Closed PnL'].agg(lambda x: (x > 0).sum() / x.count() * 100)

overall_avg_size_usd = merged_df['Size USD'].mean()
avg_size_usd_per_account = merged_df.groupby('Account')['Size USD'].mean()

print("Win Rate per Account (%):")
print(win_rate_per_account.sort_values(ascending=False).head())

print(f"\nOverall Average 'Size USD' per trade: {overall_avg_size_usd:.2f}")

print("\nAverage 'Size USD' per Account:")
print(avg_size_usd_per_account.sort_values(ascending=False).head())
```

## Calculate Number of Trades per Day

### Subtask:
See how trading activity varies day by day and visualize the patterns over time.


**Reasoning**:
I'll group by date and count up how many trades happened each day. Simple as that.

## Calculate Long/Short Ratio

### Subtask:
Check if traders prefer going long (BUY) or short (SELL), both overall and by individual account.


**Reasoning**:
Let me count up all the BUY trades versus SELL trades to see the overall directional bias.



```python
overall_trade_counts = merged_df['Side'].value_counts()
overall_buy_trades = overall_trade_counts.get('BUY', 0)
overall_sell_trades = overall_trade_counts.get('SELL', 0)

print(f"Overall BUY trades: {overall_buy_trades}")
print(f"Overall SELL trades: {overall_sell_trades}")
```

**Reasoning**:
Now I'll divide BUYs by SELLs to get the ratio (making sure we don't divide by zero).



```python
overall_long_short_ratio = overall_buy_trades / overall_sell_trades if overall_sell_trades > 0 else float('inf')
print(f"Overall Long/Short Ratio (BUY/SELL): {overall_long_short_ratio:.2f}")
```

**Reasoning**:
Let's break this down by account to see if different traders have different directional preferences. I'll handle the math carefully to avoid any division errors.



```python
trades_per_account = merged_df.groupby('Account')['Side'].value_counts().unstack(fill_value=0)
long_short_ratio_per_account = (trades_per_account['BUY'] / trades_per_account['SELL']).replace([float('inf'), -float('inf')], pd.NA)

print("Long/Short Ratio (BUY/SELL) per Account:")
print(long_short_ratio_per_account.sort_values(ascending=False).head(10))
```

# Task
Now for the interesting part - how does market sentiment affect trading results? Let's break down PnL, win rates, and losses by each sentiment category and see what patterns emerge.

## Analyze Performance Metrics by Sentiment

### Subtask:
Compare how profitable trading is under different market moods - Fear, Greed, and everything in between.


**Reasoning**:
I'll group trades by sentiment and calculate key metrics for each - total PnL, win rate, and how bad the losses are. This will give us a complete picture of performance in each market condition.



```python
sentiment_pnl = merged_df.groupby('classification')['Closed PnL'].sum()
sentiment_win_rate = merged_df.groupby('classification')['Closed PnL'].apply(lambda x: (x > 0).sum() / len(x) * 100)
sentiment_avg_losing_pnl = merged_df[merged_df['Closed PnL'] < 0].groupby('classification')['Closed PnL'].mean()

performance_by_sentiment = pd.DataFrame({
    'Total PnL': sentiment_pnl,
    'Win Rate (%)': sentiment_win_rate,
    'Average Losing PnL': sentiment_avg_losing_pnl
}).fillna(0) # Fill NaN for classifications with no losing trades or no trades

print("Performance Metrics by Sentiment Classification:")
print(performance_by_sentiment.sort_values(by='Total PnL', ascending=False))
```

## Analyze Trader Behavior by Sentiment

### Subtask:
Let's see how traders actually behave in different market conditions - do they trade more? Go long or short? Size up or down?


**Reasoning**:
Starting with trade frequency - how active are traders in each sentiment phase?



```python
sentiment_trade_frequency = merged_df.groupby('classification').size()
print("Trade Frequency by Sentiment Classification:")
print(sentiment_trade_frequency)
```

**Reasoning**:
Now let me add in the directional bias and position sizing for each sentiment. Putting it all together will show the complete behavioral picture.



```python
sentiment_trades_side = merged_df.groupby('classification')['Side'].value_counts().unstack(fill_value=0)
sentiment_long_short_ratio = (sentiment_trades_side['BUY'] / sentiment_trades_side['SELL']).replace([float('inf'), -float('inf')], pd.NA)

sentiment_avg_position_size = merged_df.groupby('classification')['Size USD'].mean()

behavior_by_sentiment = pd.DataFrame({
    'Trade Frequency': sentiment_trade_frequency,
    'Long/Short Ratio': sentiment_long_short_ratio,
    'Average Position Size (USD)': sentiment_avg_position_size
}).fillna(0) # Fill NA for classifications with no 'SELL' trades, implying an infinite ratio becomes 0 for display

print("Trader Behavior by Sentiment Classification (Sorted by Trade Frequency):")
print(behavior_by_sentiment.sort_values(by='Trade Frequency', ascending=False))
```

## Segment Traders

### Subtask:
Let's split traders into groups - who's active vs casual? Who's consistently profitable vs everyone else?


**Reasoning**:
First, I need to count total trades per account so I can separate the active traders from the occasional ones.



```python
total_trades_per_account = merged_df.groupby('Account').size().rename('Total Trades')
print("Total Trades per Account (first 5 accounts):")
print(total_trades_per_account.head())
```

**Reasoning**:
I'll use the median as a dividing line - accounts above it are 'frequent', below are 'infrequent'. Then I'll compare the two groups.



```python
trades_median = total_trades_per_account.median()
frequent_traders = total_trades_per_account[total_trades_per_account > trades_median]
infrequent_traders = total_trades_per_account[total_trades_per_account <= trades_median]

print(f"Median number of trades used as threshold: {trades_median}")

print("\nDescriptive statistics for Frequent Traders (Total Trades):")
print(frequent_traders.describe())

print("\nDescriptive statistics for Infrequent Traders (Total Trades):")
print(infrequent_traders.describe())
```

**Reasoning**:
To find consistent winners, I need both PnL and win rate data together. Let me merge those metrics.



```python
account_performance = pd.DataFrame({
    'Total PnL': total_pnl_per_account,
    'Win Rate (%)': win_rate_per_account
})

print("Account Performance (first 5 rows):")
print(account_performance.head())
```

**Reasoning**:
Consistent winners are those who make money AND have above-average win rates. Let me use the median win rate as the bar and separate the wheat from the chaff.



```python
median_win_rate = account_performance['Win Rate (%)'].median()

consistent_winners = account_performance[
    (account_performance['Total PnL'] > 0) &
    (account_performance['Win Rate (%)'] > median_win_rate)
]
inconsistent_traders = account_performance.drop(consistent_winners.index)

print(f"Median Win Rate used as threshold: {median_win_rate:.2f}%")

print("\nDescriptive statistics for Consistent Winners:")
print(consistent_winners.describe())

print("\nDescriptive statistics for Inconsistent Traders:")
print(inconsistent_traders.describe())
```
